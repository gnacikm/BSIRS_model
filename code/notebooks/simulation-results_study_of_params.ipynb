```python
def read_file_regpar(lam_val, beta):
    df_c_vals = pd.read_csv(data_path/f"saved_params/beta_{beta}_lam_{lam_val}_sim/optimal_params_c_vals_lam_is_{lam_val}.csv")
    c_array = df_c_vals.values.transpose()
    result = np.array( [np.power(W@c - c, 2) for c in c_array]).mean()
    return beta**2*result
```

```python
def read_file_score(lam_val, beta):
    df_score00 = pd.read_csv(data_path/f"saved_params/beta_{beta}_lam_{lam_val}_sim/optimal_params_score_lam_is_{lam_val}.csv")
    score_array = df_score00.values.transpose()
    return score_array.mean()
```
